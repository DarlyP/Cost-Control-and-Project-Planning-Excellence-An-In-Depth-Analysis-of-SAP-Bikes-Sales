#### **Show Row & Column Data**

---

```python
# Function shape
def shape(file, file_name):
    # Get dataframe
    rows, cols = file.shape
    
    # Return string
    return f"{file_name} has {rows} rows and {cols} columns."
```

```python
# Show row and column
print(shape(Addresses_csv, "Addresses.csv"))
print(shape(BusinessPartners_csv, "BusinessPartners.csv"))
print(shape(Employees_csv, "Employees.csv"))
print(shape(ProductCategories_csv, "ProductCategories.csv"))
print(shape(ProductCategoryText_csv, "ProductCategoryText.csv"))
print(shape(ProductTexts_csv, "ProductTexts.csv"))
print(shape(Products_csv, "Products.csv"))
print(shape(SalesOrderItems_csv, "SalesOrderItems.csv"))
print(shape(SalesOrders_csv, "SalesOrders.csv"))
```

#### **Show Types Data**

---

```python
# Function data types
def data_types(file, file_name):
    # Get type data in every column
    types = file.dtypes

    # Return string
    return f"Data types in {file_name}:\n{types}\n"
```

```python
# Show type data
print(data_types(Addresses_csv, "Addresses.csv"))
print(data_types(BusinessPartners_csv, "BusinessPartners.csv"))
print(data_types(Employees_csv, "Employees.csv"))
print(data_types(ProductCategories_csv, "ProductCategories.csv"))
print(data_types(ProductCategoryText_csv, "ProductCategoryText.csv"))
print(data_types(ProductTexts_csv, "ProductTexts.csv"))
print(data_types(Products_csv, "Products.csv"))
print(data_types(SalesOrderItems_csv, "SalesOrderItems.csv"))
print(data_types(SalesOrders_csv, "SalesOrders.csv"))
```

#### **Show Duplicate Data**

---

```python
# Function duplicated data
def duplicated(file, file_name):
    # Get type data in every column
    duplicate = file.duplicated().sum()

    # Return string
    return f"Data duplicate in {file_name}:\n{duplicate}\n"
```

```python
# Show duplicated data
print(duplicated(Addresses_csv, "Addresses.csv"))
print(duplicated(BusinessPartners_csv, "BusinessPartners.csv"))
print(duplicated(Employees_csv, "Employees.csv"))
print(duplicated(ProductCategories_csv, "ProductCategories.csv"))
print(duplicated(ProductCategoryText_csv, "ProductCategoryText.csv"))
print(duplicated(ProductTexts_csv, "ProductTexts.csv"))
print(duplicated(Products_csv, "Products.csv"))
print(duplicated(SalesOrderItems_csv, "SalesOrderItems.csv"))
print(duplicated(SalesOrders_csv, "SalesOrders.csv"))
```

---

#### **Show Missing Value Data**

---

```python
def missing_value(file, file_name):
    # Show missing value
    missing_count = file.isnull().sum()

    # Row total
    total = file.shape[0]

    # Missing value percentage
    percentage = (missing_count / total) * 100

    # Create dataframe
    missing_data = pd.DataFrame({
        'Missing Count': missing_count,
        'Percentage': percentage
    })

    # Show result
    print(f"Missing values in {file_name}:")
    print(missing_data)
    print("\n")  # Tambahkan baris kosong untuk pemisahan
```

```python
# Show duplicated data
missing_value(Addresses_csv, "Addresses.csv")
missing_value(BusinessPartners_csv, "BusinessPartners.csv")
missing_value(Employees_csv, "Employees.csv")
missing_value(ProductCategories_csv, "ProductCategories.csv")
missing_value(ProductCategoryText_csv, "ProductCategoryText.csv")
missing_value(ProductTexts_csv, "ProductTexts.csv")
missing_value(Products_csv, "Products.csv")
missing_value(SalesOrderItems_csv, "SalesOrderItems.csv")
missing_value(SalesOrders_csv, "SalesOrders.csv")
```

#### **Cleaning Missing value**

---

```python
# Functiong missing rows
def missing_rows(file):
    missing = file[file.isnull().any(axis=1)]
    return missing
```

```python
def missing_value(file, file_name):
    # Show missing value
    missing_count = file.isnull().sum()

    # Row total
    total = file.shape[0]

    # Missing value percentage
    percentage = (missing_count / total) * 100

    # Create dataframe
    missing_data = pd.DataFrame({
        'Missing Count': missing_count,
        'Percentage': percentage
    })

    # Show result
    print(f"Missing values in {file_name}:")
    print(missing_data)
    print("\n")  # Tambahkan baris kosong untuk pemisahan
```

```python
# Show duplicated data
missing_value(Addresses_csv, "Addresses.csv")
missing_value(BusinessPartners_csv, "BusinessPartners.csv")
missing_value(Employees_csv, "Employees.csv")
missing_value(ProductCategories_csv, "ProductCategories.csv")
missing_value(ProductCategoryText_csv, "ProductCategoryText.csv")
missing_value(ProductTexts_csv, "ProductTexts.csv")
missing_value(Products_csv, "Products.csv")
missing_value(SalesOrderItems_csv, "SalesOrderItems.csv")
missing_value(SalesOrders_csv, "SalesOrders.csv")
```
